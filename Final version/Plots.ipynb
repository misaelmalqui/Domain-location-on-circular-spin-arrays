**Packages**

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import locale
```

**Preamble**

```python
plt.rcParams.update({"text.usetex": True})
pre = r"""  \usepackage{amsmath}
            \usepackage{amssymb}
            \usepackage[decimalsymbol=comma]{siunitx}"""
plt.rc('text.latex', preamble=pre)

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True

def catalan(action):
    if action:
        locale.setlocale(locale.LC_NUMERIC, "ca_ES")
        plt.rcParams['axes.formatter.use_locale'] = True
    return None
```

```python
def get_values(file_path):
    f = open(file_path, 'r')
    lines = f.readlines()
    f.close()

    N_values = [float(line.split('\t')[0].strip()) for line in lines]
    P_values = [float(line.split('\t')[1].strip()) for line in lines]

    return N_values, P_values
```

**Data obtention**

```python
ns_1, ps_1 = get_values('Data-generation/D2-Prob-vs-N-interval1.txt')
ns_2, ps_2 = get_values('Data-generation/D2-Prob-vs-N-interval2.txt')
ns_3, ps_3 = get_values('Data-generation/D2-Prob-vs-N-interval3.txt')
ns_4, ps_4 = get_values('Data-generation/D2-Prob-vs-N-interval4.txt')

ns = ns_1 + ns_2 + ns_3 + ns_4
ps = ps_1 + ps_2 + ps_3 + ps_4
```

```python
print(ns)
print(ps)
```

**Plot**

```python
# List of colors: #6dbf8d, #ed8072, #118ab2
font_size=19
plt.figure(figsize=(8,5))
plt.plot(ns, ps, marker='o', color='#118ab2', label=r'$\mathrm{SRM}$')
plt.axhline(y=0.775567, color='black', linestyle='--', dashes=(5,5), label=r'$p_\mathrm{asymp}$')

plt.xlabel(r'$N$', fontsize=font_size)
plt.ylabel(r'$P_\mathrm{success}$', fontsize=font_size)

plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)

ax = plt.gca()
#ax.set_xlim([xmin, xmax])
ax.set_ylim([0.677, 0.78])

plt.legend(fontsize=font_size)
plt.savefig('Figures-generation/D2-prob-vs-N.pdf', bbox_inches='tight')
plt.show()
```
